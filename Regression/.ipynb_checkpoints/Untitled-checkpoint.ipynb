```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset: House Size (sq ft) vs Price ($1000s)
X = np.array([400, 600, 800, 1000])
Y = np.array([50, 80, 110, 140])

# Reshape X for sklearn
X_reshaped = X.reshape(-1, 1)

# Train the model
model = LinearRegression()
model.fit(X_reshaped, Y)

# Predict price for a 750 sq ft house
predicted_price = model.predict(np.array([[750]]))
print(f"Predicted price for 750 sq ft house: ${predicted_price[0]:.2f}K")

# Plotting
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, model.predict(X_reshaped), color='red', label='Best Fit Line')
plt.scatter(750, predicted_price, color='green', marker='o', label='Prediction')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price ($1000s)')
plt.legend()
plt.show()

# Compare with model parameters
print("Sklearn Slope:", model.coef_[0])
print("Sklearn Intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset: Hours studied vs Exam Score
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])

# Normalize X for better gradient descent convergence
X = (X - np.mean(X)) / np.std(X)

# Initialize parameters
m = 0  # Slope
c = 0  # Intercept
alpha = 0.01  # Learning rate
epochs = 1000  # Number of iterations
n = len(X)

# Gradient Descent
for _ in range(epochs):
    Y_pred = m * X + c  # Predicted values
    error = Y_pred - Y  # Error term
    m -= alpha * (2/n) * np.sum(error * X)  # Update slope
    c -= alpha * (2/n) * np.sum(error)  # Update intercept

# Final equation
print(f"Final Equation: Y = {m:.2f}X + {c:.2f}")

# Plotting
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, m * X + c, color='red', label='Best Fit Line')
plt.xlabel('Hours Studied (Normalized)')
plt.ylabel('Exam Score')
plt.legend()
plt.show()

# Compare with Scikit-Learn
X_reshaped = X.reshape(-1, 1)  # Reshape for sklearn
model = LinearRegression()
model.fit(X_reshaped, Y)
print("Sklearn Slope:", model.coef_[0])
print("Sklearn Intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset: House Size (sq ft), Number of Rooms vs Price ($1000s)
X = np.array([
    [400, 2],
    [600, 3],
    [800, 3],
    [1000, 4]
])
Y = np.array([50, 80, 110, 140])

# Train the model
model = LinearRegression()
model.fit(X, Y)

# Predict price for a 750 sq ft house with 3 rooms
predicted_price = model.predict(np.array([[750, 3]]))
print(f"Predicted price for 750 sq ft house with 3 rooms: ${predicted_price[0]:.2f}K")

# Compare with model parameters
print("Sklearn Coefficients (Slopes):", model.coef_)
print("Sklearn Intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset: Hours Studied, Previous Test Score vs Final Exam Score
X = np.array([
    [2, 60],
    [4, 70],
    [6, 80],
    [8, 85]
])
Y = np.array([55, 65, 75, 80])

# Train the model
model = LinearRegression()
model.fit(X, Y)

# Predict final exam score for a student who studied 7 hours and had a previous test score of 75
predicted_score = model.predict(np.array([[7, 75]]))
print(f"Predicted final exam score for 7 hours studied and 75 previous test score: {predicted_score[0]:.2f}")

# Compare with model parameters
print("Sklearn Coefficients (Slopes):", model.coef_)
print("Sklearn Intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset: Years of Experience, Education Level (in years) vs Salary ($1000s)
X = np.array([
    [1, 12],  # 1 year experience, 12 years education
    [3, 14],  # 3 years experience, 14 years education
    [5, 16],  # 5 years experience, 16 years education
    [7, 18],  # 7 years experience, 18 years education
    [10, 20]  # 10 years experience, 20 years education
])
Y = np.array([30, 50, 70, 90, 120])  # Salaries in $1000s

# Train the model
model = LinearRegression()
model.fit(X, Y)

# Predict salary for a person with 6 years experience and 16 years of education
predicted_salary = model.predict(np.array([[6, 16]]))
print(f"Predicted salary for 6 years experience and 16 years education: ${predicted_salary[0]:.2f}K")

# Compare with model parameters
print("Sklearn Coefficients (Slopes):", model.coef_)
print("Sklearn Intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset: Engine Size (L), Weight (kg) vs Fuel Efficiency (km/L)
X = np.array([
    [1.2, 800],
    [1.5, 1000],
    [2.0, 1100],
    [2.8, 1400]
])
Y = np.array([18, 15, 12, 9])  # Fuel Efficiency in km/L

# Train the model
model = LinearRegression()
model.fit(X, Y)

# Predict fuel efficiency for a car with 2.5L engine size and 1200 kg weight
predicted_efficiency = model.predict(np.array([[2.5, 1200]]))
print(f"Predicted fuel efficiency for 2.5L engine and 1200kg weight: {predicted_efficiency[0]:.2f} km/L")

# Compare with model parameters
print("Sklearn Coefficients (Slopes):", model.coef_)
print("Sklearn Intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset: Days vs Stock Price ($)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Days
Y = np.array([100, 102, 105, 107, 110])  # Stock Prices

# Train the model
model = LinearRegression()
model.fit(X, Y)

# Predict stock price on day 6
predicted_price = model.predict(np.array([[6]]))
print(f"Predicted stock price on day 6: ${predicted_price[0]:.2f}")

# Plotting
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Trend Line')
plt.scatter(6, predicted_price, color='green', marker='o', label='Prediction')
plt.xlabel('Days')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.show()

# Compare with model parameters
print("Sklearn Slope:", model.coef_[0])
print("Sklearn Intercept:", model.intercept_)
```
