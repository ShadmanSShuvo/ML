```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
# Generate synthetic non-linear data
np.random.seed(42)
X = np.linspace(-2.1, 2.1, 100).reshape(-1, 1)  # Feature
y = X*(X-1)*(X-2)*(X+1)*(X+2) + np.random.randn(100, 1)  # Quadratic relationship with noise

# Scatter plot of data
plt.scatter(X, y, label="Data")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.title("Non-Linear Data")
plt.legend()
plt.show()
```

```python
# Polynomial Regression (Degree 5)
poly_reg = make_pipeline(PolynomialFeatures(degree=5), LinearRegression())
poly_reg.fit(X, y)

# Predictions
y_pred_poly = poly_reg.predict(X)
y2 = X*(X-1)*(X-2)*(X+1)*(X+2)
# Plot results
plt.scatter(X, y, label="Data")
plt.plot(X, y_pred_poly, color="red", label="Polynomial Fit (Degree 5)")
plt.plot(X, y2, color="green", label="Original Polynomial (Degree 5)")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.title("Polynomial Regression (Degree 3)")
plt.legend()
plt.show()
```
